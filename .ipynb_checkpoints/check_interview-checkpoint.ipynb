```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

Для удобства база данных выгружена в файл csv. Важно отметить, что поиск автомобилей на сайтах auto.ru и avito.ru несколько различается. Главное отличие - выбрать автомобили по признаку "Отечественные" на auto.ru можно только из категории "Легковые", а на avito.ru только из общей категории "Автомобили". Если предположить, что данные auto.ru...

Утверждение № 1:  Средняя цена отечественного автомобиля с пробегом - около 400 тысяч рублей

```python
df = pd.read_csv("interview_data.csv", sep=";", encoding="utf-8")
df_pas = df[~(df["body_type"].isin(['Микроавтобус', 'Фургон', 'Минивэн', 'Пикап']) & (df["brand"].isin(["ГАЗ", "УАЗ"]))) | \
         ~((df["brand"].isin(["ГАЗ", "УАЗ"])) & (~df["model"].isin(["Карго", "Pickup"])))] \
         [['brand', 'model', 'car_year', 'mileage', 'body_type', 'price']]
print(df.head())
print(df.shape)
print(df_pas.shape)
```

```python
avito_mean_price = df_pas.price.mean()
round(avito_mean_price, 2)
```

Средняя цена автомобилей в нашей выборке несколько выше указанной в интервью. Проверим выбросы

```python
plt.boxplot(df_pas.price)
plt.title('Boxplot распределения цен')
plt.xlabel("Легковые автомобили")
plt.ylabel("Цена автомобиля")
```

Наблюдается достаточно большое количество выбросов, но в то же время порог отсчета довольно невысокий - 1,5 млн. Возможно эти значения не совсем корректны. Проверим максимальные 

```python
df_pas.sort_values("price", ascending=False)[["brand", "model", "price"]].head(10)
```

Основная часть самых дорогих авто - внедорожники Нива и Патриот. Судя по всему, они относятся к наиболее дорогой категории отечественных автомобилей. Самое дорогое авто - тюнингованный ГАЗ 1966 года. Такой уникальный автомобиль можно исключить без потери качества данных. Фото, если станет интересно

```python
df_pas.drop(2462, inplace=True)
avito_mean_price = df_pas.price.mean()
round(avito_mean_price, 2)
```

Учитывая размеры выборки, удаление самого крупного значения не привело к существенным изменениям. Посмотрим на распределение цен, что поможет нам также с проверкой следующего утверждения

```python
# draw histogram
plt.hist(df_pas.price, 30, density=True, facecolor='g', alpha=0.75, label='Гистограмма')
plt.title('Гистограмма распределения цен')
plt.xlabel('Цена')
plt.ylabel('Плотность наблюдений')
plt.grid(True)

mean = df_pas.price.mean()
sd = df_pas.price.std()
x = np.linspace(mean - 3 * sd, mean + 3 * sd, 1000)
plt.plot(x, stats.norm.pdf(x, mean, sd), color='red', label='Z-распределение')
```

Ожидаемо распределение цен автомобилей смещено к меньшим значениям и совершенно не похоже на нормальное. Тем не менеее, наша выборка достаточно велика (> 5000  значений), отдельные выбросы не влияют на нее сильно, и можно предположить, что она отражает распределение генеральной совокупности. Для того, чтобы убедиться, что различия значимы, расчитаем доверительный интервал с помощью Bootstrap

```python
res = stats.bootstrap((df_pas.price,), np.mean, method='basic')
is_into = 400000 > res.confidence_interval.low and 400000 < res.confidence_interval.high
conclusion = f"Средняя цена автомобиля на Авито {round(avito_mean_price, 2)} рублей. Отличие от обозначенного в интервью значения {('статистически значимо', 'не является статистически значимым')[is_into]}"
conclusion
```

Далее перейдем к утверждению утверждению № 2:
Больше половины (53%) предложения - это сегмент отечественных автомобилей с пробегом не дороже 300 тысяч рублей. Еще 16% занимают машины стоимостью 300-500 тысяч, 15% - 800 тысяч - 1,5 млн, тоже 15% - 500-800 тысяч рублей.

```python
price_dist = pd.Series({"<300": 53, "300-500": 16, "500-800": 15, "800-1500": 15}).to_frame()
price_dist["Avito"] = round(pd.cut(df_pas["price"], [0, 300_000, 500_000, 800_000, 1_500_000]).value_counts(sort=False, normalize=True) * 100, 2).values
price_dist.rename({0: "Auto.ru"}, axis=1, inplace=True)
price_dist
```

Утверждение № 3: Четверть (25%) предложения занимают старые автомобили отечественного производства - старше 20 лет. По 21% приходится на машины возрастом 15-19 лет и 10-14 лет, 14% - 3-5 лет, 7% - относительно свежие, не старше 2 лет?

```python
age_dist = pd.Series({">20": 25, "15-19": 21, "10-14": 21, "5-10": 12, "3-5": 14, "<2": 7}).to_frame()
age_dist["Avito"] = round(pd.cut((2023 - df_pas["car_year"]), [0, 2, 5, 10, 14, 19, 200], include_lowest=True).value_counts(sort=False, normalize=True)[::-1] * 100, 2).values
age_dist.rename({0: "Auto.ru"}, axis=1, inplace=True)
age_dist
```

Рапределение автомобилей по годам различается лишь в двух группах: На Авито больше объявлений в промежутке от 5 до 10 лет и меньше объявлений возрастных автомобилей старше 20 лет. 
Утверждение № 4: Средний возраст отечественного автомобиля с пробегом, - 14 лет

```python
avito_mean_age = (2023 - df_pas.car_year).mean()
res = stats.ttest_1samp(df.price, 13.51)
conclusion = f"Средний возраст автомобиля на Авито {round(avito_mean_age, 2)} лет. Отличие от обозначенного в интервью значения {('статистически значимо', 'не является статистически значимым')[res.pvalue>0.05]}"
conclusion
```
