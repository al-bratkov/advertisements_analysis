Летом 2023 года был составлен датасет, содержащий данные об объявлениях о продаже автомобилей на сайте avito.ru, по следующим критериям:
- тип автомобиля: С пробегом,
- производитель: Отечественные,
- состояние: Кроме битых,
- отметка "Частные",
- отметка "Собственники и проверенные партнеры".

Полученные данные интересно проверить. Но проблема в том, что не ясно как получить истинные значения, которые можно было бы взять за образец. Для сравнения мы будем использовать данные из интервью руководителя группы исследований и клиентской аналитики "Авто.ру" Алексея Муханова, описывающем рынок поддержанных отечественных автомобилей. Интервью вышло 23 мая 2023г, а данные отбирались летом. Из-за этого могут возникнуть расхождения, но, к сожалению, более подходящей информации не нашлось. 
Рассмотрим озвученные в интервью данные:
Средняя цена отечественного автомобиля с пробегом, по данным "Авто.ру", - около 400 тысяч рублей
Четверть (25%) предложения занимают старые автомобили отечественного производства - старше 20 лет. По 21% приходится на машины возрастом 15-19 лет и 10-14 лет, 14% - 3-5 лет, 7% - относительно свежие, не старше 2 лет.
Больше половины (53%) предложения - это сегмент отечественных автомобилей с пробегом не дороже 300 тысяч рублей. Еще 16% занимают машины стоимостью 300-500 тысяч, 15% - 800 тысяч - 1,5 млн, тоже 15% - 500-800 тысяч рублей.
Средний возраст отечественного автомобиля с пробегом, - 14 лет
средневзвешенный пробег достигает 134 тысячи километров.
На вторичном рынке значимую долю предложения составляют продукция АвтоВАЗа классического семейства, а также "Самара" первого и второго поколения, несмотря на то, что эти машины давно сняты с производства.
